```python
import pandas as pd
```

```python
df_full = pd.read_csv('../results/literature_review_fulldata.csv')
```

```python
df_year = pd.read_csv('../results/final_set_with_year_and_type.csv')
```

```python
df_year_selected = df_year[['doi', 'publication_year', 'cited_by_count', 'type']]
```

```python
df_year_merged = pd.merge(df_year_selected, df_full, how='outer', on='doi')
```

```python
df_unique = df_year_merged.drop_duplicates(subset=['doi', 'title'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

df = df_unique.copy()

# Drop rows with missing publication_year
df = df.dropna(subset=['publication_year']).copy()
df['publication_year'] = df['publication_year'].astype(int)

# Group by issue_area and publication_year, count articles
counts = df.groupby(['issue_area', 'publication_year']).size().unstack(fill_value=0)

# Sort issue_areas by total count descending
total_per_issue_area = counts.sum(axis=1)
counts = counts.loc[total_per_issue_area.sort_values(ascending=False).index]

# Prepare blue colors—one shade per year
years = counts.columns.tolist()
num_years = len(years)
cmap = plt.get_cmap('Blues')
colors = [cmap(0.3 + 0.7 * i / max(num_years-1, 1)) for i in range(num_years)]

# Set Times New Roman font globally and increase font sizes
plt.rcParams.update({
    'font.family': 'Times New Roman',
    'axes.titlesize': 20,      # bigger title
    'axes.labelsize': 18,      # bigger axis labels
    'xtick.labelsize': 16,     # bigger x-tick labels
    'ytick.labelsize': 16,     # bigger y-tick labels
    'legend.fontsize': 16      # bigger legend font
})

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(14, 8))

bottoms = np.zeros(len(counts))
x = np.arange(len(counts))

for i, year in enumerate(years):
    ax.bar(x, counts[year], bottom=bottoms, color=colors[i], label=str(year))
    bottoms += counts[year]

# Add total counts on top of each bar, increased font size and bold
for idx, total in enumerate(total_per_issue_area.loc[counts.index]):
    ax.text(x[idx], total + 0.5, str(total), ha='center', va='bottom', fontsize=16, fontweight='bold', family='Times New Roman')

# Axis labels and title
ax.set_xticks(x)
ax.set_xticklabels(counts.index, rotation=45, ha='right')
ax.set_ylabel('Total Counts')
ax.set_title('Total Counts per Issue Area by Publication Year (Sorted)')
ax.grid(axis='y')

# Legend with larger markers and title font
legend = ax.legend(title='Publication Year', title_fontsize='18', loc='upper right', borderpad=1)
for text in legend.get_texts():
    text.set_fontsize(16)

plt.tight_layout()

# Save figure as SVG
fig.savefig('issue_area_total_counts.svg', format='svg')

plt.show()
```
