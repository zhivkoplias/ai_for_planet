```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from typing import List, Optional, Tuple, Dict

def plot_issue_area_means_for_ppt(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    acc_bins: Optional[List[float]] = None,
    acc_labels: List[str] = ['Low', 'Medium', 'High'],
    size_map: Optional[Dict[str, int]] = None,
    palette: Optional[List] = None,
    annotate: bool = True,
    title_fontsize: int = 22,
    label_fontsize: int = 18,
    tick_fontsize: int = 16,
    legend_fontsize: int = 16,
    annotation_fontsize: int = 16,
    legend_marker_scale: float = 5,   # controls legend marker appearance only
    legend_box_width: float = 0.30
) -> Tuple[pd.DataFrame, plt.Figure, plt.Axes]:
    """
    Plot issue-area means with larger fonts & enlarged legend suitable for PPT.
    - Urgency parameter removed entirely.
    - All markers use the same fixed `point_size`.
    - Accessibility bins default to arbitrary thresholds: Low < 2, 2-3 = Medium, >3 High.
    Returns (summary DataFrame, fig, ax).
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    if palette is None:
        palette = sns.color_palette('Set1', n_colors=len(acc_labels))

    if size_map is None:
        size_map = {'Low': 160, 'Medium': 480, 'High': 1200}

    # Build summary (only data, performance, accessibility)
    rows = []
    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'

        if not {col_data, col_perf, col_acc}.issubset(df.columns):
            continue

        rows.append({
            'issue': issue,
            'mean_data': pd.to_numeric(df[col_data], errors='coerce').mean(skipna=True),
            'mean_performance': pd.to_numeric(df[col_perf], errors='coerce').mean(skipna=True),
            'mean_accessibility': pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True)
        })

    summary = pd.DataFrame(rows)
    if summary.empty:
        raise ValueError("No issue areas with required columns (data, performance, accessibility) found in DataFrame.")

    # Binning accessibility (use arbitrary thresholds if acc_bins not supplied)
    if acc_bins is None:
        def acc_bin_from_value(x):
            if pd.isna(x):
                return np.nan
            try:
                xv = float(x)
            except Exception:
                return np.nan
            if xv < 2.0:
                return acc_labels[0]   # 'Low'
            elif xv <= 3.0:
                return acc_labels[1]   # 'Medium'
            else:
                return acc_labels[2]   # 'High'
        summary['acc_bin'] = summary['mean_accessibility'].apply(acc_bin_from_value).astype('category')
    else:
        try:
            summary['acc_bin'] = pd.cut(summary['mean_accessibility'], bins=acc_bins, labels=acc_labels, include_lowest=True)
        except Exception:
            summary['acc_bin'] = pd.qcut(summary['mean_accessibility'], q=3, labels=acc_labels)

    # Convert categorical bins to plain strings (or NaN)
    summary['acc_bin_str'] = summary['acc_bin'].astype(object)
    summary.loc[summary['acc_bin'].isna(), 'acc_bin_str'] = np.nan
    summary['acc_bin_str'] = summary['acc_bin_str'].astype('object')

    # Map to colors safely
    color_map = dict(zip(list(map(str, acc_labels)), palette))
    mapped_colors = summary['acc_bin_str'].astype(str).map(color_map)
    mapped_colors = mapped_colors.where(~mapped_colors.isnull(), other=np.nan)
    fallback_color = (0.7, 0.7, 0.7)
    summary['color'] = mapped_colors.apply(lambda v: v if (not (v is np.nan or pd.isna(v))) else fallback_color)

    # Ensure equal dot sizes for every point
    point_size = 500.0   # area in points^2 used by matplotlib scatter's `s` argument
    summary['size'] = point_size

    # Prepare figure and large fonts
    FIXED_SIZE_INCHES = (16, 10)
    fig, ax = plt.subplots()
    fig.set_size_inches(*FIXED_SIZE_INCHES, forward=True)
    sns.set(style='whitegrid')

    plt.rcParams.update({
        'font.size': tick_fontsize,
        'axes.titlesize': title_fontsize,
        'axes.labelsize': label_fontsize,
        'xtick.labelsize': tick_fontsize,
        'ytick.labelsize': tick_fontsize,
        'legend.fontsize': legend_fontsize
    })

    # Plot points — use constant size
    for _, r in summary.iterrows():
        ax.scatter(r['mean_data'], r['mean_performance'],
                   s=float(r['size']),
                   color=r['color'],
                   edgecolor='k',
                   linewidth=0.9,
                   alpha=0.95)
        if annotate:
            ax.text(r['mean_data'], r['mean_performance'], ' ' + r['issue'],
                    verticalalignment='center', fontsize=annotation_fontsize)

    # Build legend handles for accessibility (colors only) with custom legend title and labels including boundaries
    color_handles = []
    # markersize in legend is diameter in points; scale up for PPT visibility
    color_marker_size = 20 * legend_marker_scale

    # Build explicit labels showing boundaries
    label_map_with_bounds = {
        'Low':  'Low (< 2)',
        'Medium': 'Medium (2–3)',
        'High': 'High (> 3)'
    }

    for lab in acc_labels:
        lab_str = str(lab)
        if lab_str in summary['acc_bin_str'].astype(str).values:
            display_label = label_map_with_bounds.get(lab_str, lab_str)
            color_handles.append(Line2D([0], [0], marker='o', color='w',
                                       markerfacecolor=color_map[lab_str],
                                       markeredgecolor='k', markersize=color_marker_size,
                                       markeredgewidth=1.2, label=display_label))

    if color_handles:
        legend_x = 1.02
        legend_y = 0.5
        # set legend title to "Accessibility for non-expert"
        leg = ax.legend(handles=color_handles, title="Accessibility for non-experts",
                        bbox_to_anchor=(legend_x, legend_y),
                        loc='center left', frameon=True, borderpad=1.2,
                        labelspacing=1.2, handlelength=1.5)
        # set legend title fontsize and text fontsize
        if leg.get_title() is not None:
            leg.get_title().set_fontsize(legend_fontsize + 1)
        for text in leg.get_texts():
            text.set_fontsize(legend_fontsize)

    # Labels, title and grid
    ax.set_xlabel('Mean score of data access and quality', fontsize=label_fontsize)
    ax.set_ylabel('Mean score of model performance', fontsize=label_fontsize)
    ax.set_title('', fontsize=title_fontsize)
    ax.grid(alpha=0.25)

    # Make axes spines a bit thicker for PPT visibility
    for spine in ax.spines.values():
        spine.set_linewidth(1.0)

    # Reserve space on the right for the legend so it won't be clipped.
    right_margin = max(0.80, 1.0 - legend_box_width)
    fig.tight_layout(rect=[0, 0, right_margin, 1.0])

    return summary, fig, ax
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from typing import List, Optional, Tuple, Dict

def plot_issue_area_means_for_ppt(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    acc_bins: Optional[List[float]] = None,
    acc_labels: List[str] = ['Low', 'Medium', 'High'],
    size_map: Optional[Dict[str, int]] = None,
    palette: Optional[List] = None,
    annotate: bool = True,
    title_fontsize: int = 24,
    label_fontsize: int = 20,
    tick_fontsize: int = 18,
    legend_fontsize: int = 18,
    annotation_fontsize: int = 18,
    legend_marker_scale: float = 5,   # controls legend marker appearance only
    legend_box_width: float = 0.30,
    # New parameters for arrow/annotation control
    annotation_dx: float = 0.08,   # fraction of x-axis span to offset annotation in x
    annotation_dy: float = 0.0,    # fraction of y-axis span to offset annotation in y
    annotation_arrowprops: Optional[dict] = None,
    connect_points: bool = True,
    connect_line_style: dict = None
) -> Tuple[pd.DataFrame, plt.Figure, plt.Axes]:
    """
    Plot issue-area means with larger fonts & enlarged legend suitable for PPT,
    plus arrows connecting annotations to their markers and (optionally) arrows
    connecting consecutive points.

    Returns (summary DataFrame, fig, ax).
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    if palette is None:
        palette = sns.color_palette('Set1', n_colors=len(acc_labels))

    if size_map is None:
        size_map = {'Low': 160, 'Medium': 480, 'High': 1200}

    # Build summary (only data, performance, accessibility)
    rows = []
    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'

        if not {col_data, col_perf, col_acc}.issubset(df.columns):
            continue

        rows.append({
            'issue': issue,
            'mean_data': pd.to_numeric(df[col_data], errors='coerce').mean(skipna=True),
            'mean_performance': pd.to_numeric(df[col_perf], errors='coerce').mean(skipna=True),
            'mean_accessibility': pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True)
        })

    summary = pd.DataFrame(rows)
    if summary.empty:
        raise ValueError("No issue areas with required columns (data, performance, accessibility) found in DataFrame.")

    # Binning accessibility (use arbitrary thresholds if acc_bins not supplied)
    if acc_bins is None:
        def acc_bin_from_value(x):
            if pd.isna(x):
                return np.nan
            try:
                xv = float(x)
            except Exception:
                return np.nan
            if xv < 2.0:
                return acc_labels[0]   # 'Low'
            elif xv <= 3.0:
                return acc_labels[1]   # 'Medium'
            else:
                return acc_labels[2]   # 'High'
        summary['acc_bin'] = summary['mean_accessibility'].apply(acc_bin_from_value).astype('category')
    else:
        try:
            summary['acc_bin'] = pd.cut(summary['mean_accessibility'], bins=acc_bins, labels=acc_labels, include_lowest=True)
        except Exception:
            summary['acc_bin'] = pd.qcut(summary['mean_accessibility'], q=3, labels=acc_labels)

    # Convert categorical bins to plain strings (or NaN)
    summary['acc_bin_str'] = summary['acc_bin'].astype(object)
    summary.loc[summary['acc_bin'].isna(), 'acc_bin_str'] = np.nan
    summary['acc_bin_str'] = summary['acc_bin_str'].astype('object')

    # Map to colors safely
    color_map = dict(zip(list(map(str, acc_labels)), palette))
    mapped_colors = summary['acc_bin_str'].astype(str).map(color_map)
    mapped_colors = mapped_colors.where(~mapped_colors.isnull(), other=np.nan)
    fallback_color = (0.7, 0.7, 0.7)
    summary['color'] = mapped_colors.apply(lambda v: v if (not (v is np.nan or pd.isna(v))) else fallback_color)

    # Ensure equal dot sizes for every point
    point_size = 500.0   # area in points^2 used by matplotlib scatter's `s` argument
    summary['size'] = point_size

    # Prepare figure and large fonts
    FIXED_SIZE_INCHES = (16, 10)
    fig, ax = plt.subplots()
    fig.set_size_inches(*FIXED_SIZE_INCHES, forward=True)
    sns.set(style='whitegrid')

    plt.rcParams.update({
        'font.size': tick_fontsize,
        'axes.titlesize': title_fontsize,
        'axes.labelsize': label_fontsize,
        'xtick.labelsize': tick_fontsize,
        'ytick.labelsize': tick_fontsize,
        'legend.fontsize': legend_fontsize
    })

    # Defaults for arrowprops if not provided
    if annotation_arrowprops is None:
        annotation_arrowprops = dict(arrowstyle='->', color='black', linewidth=0.9, shrinkA=0, shrinkB=2)
    if connect_line_style is None:
        connect_line_style = dict(color='gray', linestyle='-', linewidth=1.0, alpha=0.6, arrowstyle='-|>', mutation_scale=10)

    # Plot points — use constant size
    x_vals = summary['mean_data'].values
    y_vals = summary['mean_performance'].values

    # scatter all points (so z-order consistent) and store artists
    for _, r in summary.iterrows():
        ax.scatter(r['mean_data'], r['mean_performance'],
                   s=float(r['size']),
                   color=r['color'],
                   edgecolor='k',
                   linewidth=0.9,
                   alpha=0.95, zorder=3)

    # Compute axis spans to convert fraction offsets to data units
    x_min, x_max = ax.get_xlim()
    y_min, y_max = ax.get_ylim()
    # If autoscale hasn't been triggered yet (limits are default), recompute with the plotted points
    if np.isclose(x_min, x_max):
        x_min, x_max = np.min(x_vals), np.max(x_vals)
    if np.isclose(y_min, y_max):
        y_min, y_max = np.min(y_vals), np.max(y_vals)

    x_span = x_max - x_min if (x_max - x_min) != 0 else 1.0
    y_span = y_max - y_min if (y_max - y_min) != 0 else 1.0

    dx_data = annotation_dx * x_span
    dy_data = annotation_dy * y_span

    # Draw connecting arrows between consecutive points (in DataFrame order) if requested
    if connect_points and len(summary) > 1:
        # Use FancyArrowPatch for nicer arrows
        from matplotlib.patches import FancyArrowPatch
        for i in range(len(summary) - 1):
            x0, y0 = summary.iloc[i]['mean_data'], summary.iloc[i]['mean_performance']
            x1, y1 = summary.iloc[i+1]['mean_data'], summary.iloc[i+1]['mean_performance']
            # Slight offset so the arrow does not overlap with markers: shorten by a small fraction toward each end
            frac = 0.08
            sx = x0 + (x1 - x0) * frac
            sy = y0 + (y1 - y0) * frac
            ex = x1 - (x1 - x0) * frac
            ey = y1 - (y1 - y0) * frac
            arrow = FancyArrowPatch((sx, sy), (ex, ey),
                                    arrowstyle=connect_line_style.get('arrowstyle', '-|>'),
                                    mutation_scale=connect_line_style.get('mutation_scale', 12),
                                    linewidth=connect_line_style.get('linewidth', 1.0),
                                    linestyle=connect_line_style.get('linestyle', '-'),
                                    color=connect_line_style.get('color', 'gray'),
                                    alpha=connect_line_style.get('alpha', 0.6),
                                    zorder=2)
            ax.add_patch(arrow)

    # Add annotations with arrows pointing to the markers
    if annotate:
        for _, r in summary.iterrows():
            x = r['mean_data']
            y = r['mean_performance']
            # Default offset to the right; you can adjust per-point if needed
            ann_x = x + dx_data
            ann_y = y + dy_data
            # Use annotate with arrowprops so each label is connected with an arrow
            ax.annotate(text=' ' + r['issue'],
                        xy=(x, y),
                        xytext=(ann_x, ann_y),
                        fontsize=annotation_fontsize,
                        va='center',
                        ha='left' if annotation_dx >= 0 else 'right',
                        arrowprops=annotation_arrowprops,
                        bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="none", alpha=0.0),
                        zorder=4)

    # Build legend handles for accessibility (colors only) with custom legend title and labels including boundaries
    color_handles = []
    # markersize in legend is diameter in points; scale up for PPT visibility
    color_marker_size = 20 * legend_marker_scale

    # Build explicit labels showing boundaries
    label_map_with_bounds = {
        'Low':  'Low (< 2)',
        'Medium': 'Medium (2–3)',
        'High': 'High (> 3)'
    }

    for lab in acc_labels:
        lab_str = str(lab)
        if lab_str in summary['acc_bin_str'].astype(str).values:
            display_label = label_map_with_bounds.get(lab_str, lab_str)
            color_handles.append(Line2D([0], [0], marker='o', color='w',
                                       markerfacecolor=color_map[lab_str],
                                       markeredgecolor='k', markersize=color_marker_size,
                                       markeredgewidth=1.2, label=display_label))

    if color_handles:
        legend_x = 1.02
        legend_y = 0.5
        # set legend title to "Accessibility for non-expert"
        leg = ax.legend(handles=color_handles, title="Accessibility for non-experts",
                        bbox_to_anchor=(legend_x, legend_y),
                        loc='center left', frameon=True, borderpad=1.2,
                        labelspacing=1.2, handlelength=1.5)
        # set legend title fontsize and text fontsize
        if leg.get_title() is not None:
            leg.get_title().set_fontsize(legend_fontsize + 1)
        for text in leg.get_texts():
            text.set_fontsize(legend_fontsize)

    # Labels, title and grid
    ax.set_xlabel('Mean score of data access and quality', fontsize=label_fontsize)
    ax.set_ylabel('Mean score of model performance', fontsize=label_fontsize)
    ax.set_title('', fontsize=title_fontsize)
    ax.grid(alpha=0.25)

    # Make axes spines a bit thicker for PPT visibility
    for spine in ax.spines.values():
        spine.set_linewidth(1.0)

    # Reserve space on the right for the legend so it won't be clipped.
    right_margin = max(0.80, 1.0 - legend_box_width)
    fig.tight_layout(rect=[0, 0, right_margin, 1.0])

    return summary, fig, ax
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from typing import List, Optional, Tuple, Dict

def plot_issue_area_means_for_ppt(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    acc_bins: Optional[List[float]] = None,
    acc_labels: List[str] = ['Low', 'Medium', 'High'],
    size_map: Optional[Dict[str, int]] = None,
    palette: Optional[List] = None,
    annotate: bool = True,
    title_fontsize: int = 28,
    label_fontsize: int = 24,
    tick_fontsize: int = 20,
    legend_fontsize: int = 20,
    annotation_fontsize: int = 18,
    legend_marker_scale: float = 6,   # controls legend marker appearance only
    legend_box_width: float = 0.30,
    annotation_dx: float = 0.08,
    annotation_dy: float = 0.0,
    annotation_arrowprops: Optional[dict] = None,
    connect_points: bool = True,
    connect_line_style: dict = None
) -> Tuple[pd.DataFrame, plt.Figure, plt.Axes]:
    """
    Plot issue-area means with Times New Roman font and larger visual sizes
    so the figure is suitable for scientific publication / PPT.

    Returns (summary DataFrame, fig, ax).
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    if palette is None:
        palette = sns.color_palette('Set1', n_colors=len(acc_labels))

    if size_map is None:
        size_map = {'Low': 160, 'Medium': 480, 'High': 1200}

    # Build summary (only data, performance, accessibility)
    rows = []
    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'

        if not {col_data, col_perf, col_acc}.issubset(df.columns):
            continue

        rows.append({
            'issue': issue,
            'mean_data': pd.to_numeric(df[col_data], errors='coerce').mean(skipna=True),
            'mean_performance': pd.to_numeric(df[col_perf], errors='coerce').mean(skipna=True),
            'mean_accessibility': pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True)
        })

    summary = pd.DataFrame(rows)
    if summary.empty:
        raise ValueError("No issue areas with required columns (data, performance, accessibility) found in DataFrame.")

    # Binning accessibility (use arbitrary thresholds if acc_bins not supplied)
    if acc_bins is None:
        def acc_bin_from_value(x):
            if pd.isna(x):
                return np.nan
            try:
                xv = float(x)
            except Exception:
                return np.nan
            if xv < 2.0:
                return acc_labels[0]   # 'Low'
            elif xv <= 3.0:
                return acc_labels[1]   # 'Medium'
            else:
                return acc_labels[2]   # 'High'
        summary['acc_bin'] = summary['mean_accessibility'].apply(acc_bin_from_value).astype('category')
    else:
        try:
            summary['acc_bin'] = pd.cut(summary['mean_accessibility'], bins=acc_bins, labels=acc_labels, include_lowest=True)
        except Exception:
            summary['acc_bin'] = pd.qcut(summary['mean_accessibility'], q=3, labels=acc_labels)

    summary['acc_bin_str'] = summary['acc_bin'].astype(object)
    summary.loc[summary['acc_bin'].isna(), 'acc_bin_str'] = np.nan
    summary['acc_bin_str'] = summary['acc_bin_str'].astype('object')

    # Map to colors safely
    color_map = dict(zip(list(map(str, acc_labels)), palette))
    mapped_colors = summary['acc_bin_str'].astype(str).map(color_map)
    mapped_colors = mapped_colors.where(~mapped_colors.isnull(), other=np.nan)
    fallback_color = (0.7, 0.7, 0.7)
    summary['color'] = mapped_colors.apply(lambda v: v if (not (v is np.nan or pd.isna(v))) else fallback_color)

    # Ensure equal dot sizes for every point (increase for publication)
    point_size = 700.0   # area in points^2 used by matplotlib scatter's `s` argument
    summary['size'] = point_size

    # Prepare figure and large fonts
    FIXED_SIZE_INCHES = (16, 10)  # publication-friendly aspect ratio
    fig, ax = plt.subplots(figsize=FIXED_SIZE_INCHES)
    sns.set(style='whitegrid')

    # Set Times / Times New Roman as preferred serif font (fallbacks included)
    plt.rcParams.update({
        'font.family': 'serif',
        'font.serif': ['Times New Roman', 'Times', 'TimesRoman', 'DejaVu Serif'],
        'mathtext.fontset': 'stix',  # math text matches Times-like fonts
        'font.size': tick_fontsize,
        'axes.titlesize': title_fontsize,
        'axes.labelsize': label_fontsize,
        'xtick.labelsize': tick_fontsize,
        'ytick.labelsize': tick_fontsize,
        'legend.fontsize': legend_fontsize,
        'axes.linewidth': 1.0,
        'lines.linewidth': 1.5
    })

    # Defaults for arrowprops if not provided
    if annotation_arrowprops is None:
        annotation_arrowprops = dict(arrowstyle='->', color='black', linewidth=1.0, shrinkA=0, shrinkB=2)
    if connect_line_style is None:
        connect_line_style = dict(color='gray', linestyle='-', linewidth=1.0, alpha=0.6, arrowstyle='-|>', mutation_scale=12)

    # Plot points — use constant size
    x_vals = summary['mean_data'].values
    y_vals = summary['mean_performance'].values

    # scatter all points (so z-order consistent) and store artists
    for _, r in summary.iterrows():
        ax.scatter(r['mean_data'], r['mean_performance'],
                   s=float(r['size']),
                   color=r['color'],
                   edgecolor='k',
                   linewidth=1.2,
                   alpha=0.95, zorder=3)

    # Compute axis spans to convert fraction offsets to data units
    x_min, x_max = ax.get_xlim()
    y_min, y_max = ax.get_ylim()
    if np.isclose(x_min, x_max):
        x_min, x_max = np.min(x_vals), np.max(x_vals)
    if np.isclose(y_min, y_max):
        y_min, y_max = np.min(y_vals), np.max(y_vals)

    x_span = x_max - x_min if (x_max - x_min) != 0 else 1.0
    y_span = y_max - y_min if (y_max - y_min) != 0 else 1.0

    dx_data = annotation_dx * x_span
    dy_data = annotation_dy * y_span

    # Add annotations with arrows pointing to the markers
    if annotate:
        for _, r in summary.iterrows():
            x = r['mean_data']
            y = r['mean_performance']
            ann_x = x + dx_data
            ann_y = y + dy_data
            ax.annotate(text=' ' + r['issue'],
                        xy=(x, y),
                        xytext=(ann_x, ann_y),
                        fontsize=annotation_fontsize,
                        va='center',
                        ha='left' if annotation_dx >= 0 else 'right',
                        arrowprops=annotation_arrowprops,
                        bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="none", alpha=0.0),
                        zorder=4)

    # Build legend handles for accessibility (colors only)
    color_handles = []
    color_marker_size = 16 * legend_marker_scale  # markersize is diameter in points

    label_map_with_bounds = {
        'Low':  'Low (< 2)',
        'Medium': 'Medium (2–3)',
        'High': 'High (> 3)'
    }

    for lab in acc_labels:
        lab_str = str(lab)
        if lab_str in summary['acc_bin_str'].astype(str).values:
            display_label = label_map_with_bounds.get(lab_str, lab_str)
            color_handles.append(Line2D([0], [0], marker='o', color='w',
                                       markerfacecolor=color_map[lab_str],
                                       markeredgecolor='k', markersize=color_marker_size,
                                       markeredgewidth=1.4, label=display_label))

    if color_handles:
        legend_x = 1.02
        legend_y = 0.5
        leg = ax.legend(handles=color_handles, title="Accessibility for non-experts",
                        bbox_to_anchor=(legend_x, legend_y),
                        loc='center left', frameon=True, borderpad=1.2,
                        labelspacing=1.2, handlelength=1.5)
        if leg.get_title() is not None:
            leg.get_title().set_fontsize(legend_fontsize + 1)
            leg.get_title().set_fontfamily('serif')
        for text in leg.get_texts():
            text.set_fontsize(legend_fontsize)
            text.set_fontfamily('serif')

    # Labels, title and grid
    ax.set_xlabel('Mean score of data access and quality', fontsize=label_fontsize, fontfamily='serif')
    ax.set_ylabel('Mean score of model performance', fontsize=label_fontsize, fontfamily='serif')
    ax.set_title('', fontsize=title_fontsize, fontfamily='serif')
    ax.grid(alpha=0.25)

    # Make axes spines a bit thicker for publication visibility
    for spine in ax.spines.values():
        spine.set_linewidth(1.2)

    # Reserve space on the right for the legend so it won't be clipped.
    right_margin = max(0.78, 1.0 - legend_box_width)
    fig.tight_layout(rect=[0, 0, right_margin, 1.0])

    return summary, fig, ax
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Optional, Tuple, Dict

def plot_issue_distributions(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    acc_bins: Optional[List[float]] = None,
    urg_bins: Optional[List[float]] = None,
    acc_labels: List[str] = ['Low','Medium','High'],
    urg_labels: List[str] = ['Low','Medium','High'],
    figsize: Tuple[int,int] = (14,10),
    palette: Optional[List] = None,
    show_violin: bool = True,
    jitter: float = 0.12
) -> Tuple[pd.DataFrame, plt.Figure, np.ndarray]:
    """
    Plot 4 subplots (2x2) showing distributions across issue areas for:
      - data (top-left)
      - performance (top-right)
      - accessibility (bottom-left)  <-- shows dotted bin separators
      - urgency (bottom-right)        <-- shows dotted bin separators

    Parameters:
    - df: input DataFrame
    - issues: list of issue prefixes (default 8 common issues)
    - acc_bins / urg_bins: optional list of bin edges (e.g., [0,2,4,6]) used for accessibility/urgency thresholds.
                           If None, quantile-based bins (pd.qcut with q=3) are used and returned thresholds are derived from quantiles.
    - acc_labels / urg_labels: labels for bins
    - figsize: figure size
    - palette: color palette for plots (seaborn palette or list of colors). If None uses 'tab10'.
    - show_violin: whether to draw a violin behind the strip/box for visual density
    - jitter: horizontal jitter for stripplot points

    Returns:
    - summary: DataFrame with per-issue means and bin labels
    - fig, axes: matplotlib figure and axes array
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    # choose palette
    if palette is None:
        palette = sns.color_palette('tab10', n_colors=len(issues))

    # Gather long-form data for plotting
    long_rows = []
    rows = []
    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'
        col_urg  = f'{issue}.urgency'
        # if missing, skip
        if not {col_data, col_perf, col_acc, col_urg}.issubset(df.columns):
            continue

        # per-row: gather values for long-form plot (keeping issue label)
        for v in pd.to_numeric(df[col_data], errors='coerce'):
            long_rows.append({'issue': issue, 'metric': 'data', 'value': v})
        for v in pd.to_numeric(df[col_perf], errors='coerce'):
            long_rows.append({'issue': issue, 'metric': 'performance', 'value': v})
        for v in pd.to_numeric(df[col_acc], errors='coerce'):
            long_rows.append({'issue': issue, 'metric': 'accessibility', 'value': v})
        for v in pd.to_numeric(df[col_urg], errors='coerce'):
            long_rows.append({'issue': issue, 'metric': 'urgency', 'value': v})

        # compute means for summary and potential binning
        rows.append({
            'issue': issue,
            'mean_data': pd.to_numeric(df[col_data], errors='coerce').mean(skipna=True),
            'mean_performance': pd.to_numeric(df[col_perf], errors='coerce').mean(skipna=True),
            'mean_accessibility': pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True),
            'mean_urgency': pd.to_numeric(df[col_urg], errors='coerce').mean(skipna=True)
        })

    long_df = pd.DataFrame(long_rows)
    summary = pd.DataFrame(rows)

    if summary.empty:
        raise ValueError("No issue areas with all required columns present.")

    # Determine accessibility bin thresholds
    if acc_bins is None:
        # use quantiles on summary means
        # if there are fewer unique values than 3, fall back to simple min/mean/max thresholds
        try:
            q = summary['mean_accessibility'].quantile([0, 1/3, 2/3, 1.0]).values
            # ensure strictly increasing edges (if equal values exist, adjust slightly)
            acc_bins_calc = np.unique(q)
            if len(acc_bins_calc) < 4:
                # fallback to min/mean/max
                acc_bins = [summary['mean_accessibility'].min() - 1e-6,
                            summary['mean_accessibility'].median(),
                            summary['mean_accessibility'].mean(),
                            summary['mean_accessibility'].max() + 1e-6]
            else:
                acc_bins = q.tolist()
        except Exception:
            acc_bins = [summary['mean_accessibility'].min() - 1e-6,
                        summary['mean_accessibility'].median(),
                        summary['mean_accessibility'].mean(),
                        summary['mean_accessibility'].max() + 1e-6]

    # Determine urgency bin thresholds
    if urg_bins is None:
        try:
            q = summary['mean_urgency'].quantile([0, 1/3, 2/3, 1.0]).values
            urg_bins_calc = np.unique(q)
            if len(urg_bins_calc) < 4:
                urg_bins = [summary['mean_urgency'].min() - 1e-6,
                            summary['mean_urgency'].median(),
                            summary['mean_urgency'].mean(),
                            summary['mean_urgency'].max() + 1e-6]
            else:
                urg_bins = q.tolist()
        except Exception:
            urg_bins = [summary['mean_urgency'].min() - 1e-6,
                        summary['mean_urgency'].median(),
                        summary['mean_urgency'].mean(),
                        summary['mean_urgency'].max() + 1e-6]

    # Binning summary means (so we can show categories)
    summary['acc_bin'] = pd.cut(summary['mean_accessibility'], bins=acc_bins, labels=acc_labels, include_lowest=True)
    summary['urg_bin'] = pd.cut(summary['mean_urgency'], bins=urg_bins, labels=urg_labels, include_lowest=True)

    # Prepare plot layout
    fig, axes = plt.subplots(2, 2, figsize=figsize)
    axes = axes.flatten()
    sns.set(style='whitegrid')

    # order of issues (to keep color mapping consistent)
    issue_order = summary['issue'].tolist()

    # Helper plotting function: violin + strip + box
    def plot_metric(ax, metric_name, title):
        sub = long_df[long_df['metric'] == metric_name].copy()
        # Keep categorical order
        sub['issue'] = pd.Categorical(sub['issue'], categories=issue_order, ordered=True)

        # violin
        if show_violin:
            sns.violinplot(data=sub, x='issue', y='value', ax=ax, order=issue_order,
                           inner=None, palette=palette, cut=0, linewidth=0.8)
        # box (semi-transparent white boxes so violin is still visible)
        sns.boxplot(data=sub, x='issue', y='value', ax=ax, order=issue_order,
                    whis=[5,95], width=0.12, showcaps=True,
                    boxprops={'facecolor':'white','alpha':0.6}, showfliers=False)

        # strip (individual points) jittered
        sns.stripplot(data=sub, x='issue', y='value', ax=ax, order=issue_order,
                      color='k', size=4, jitter=jitter, alpha=0.7)

        ax.set_title(title)
        ax.set_xlabel('')
        ax.set_xticklabels(issue_order, rotation=45, ha='right')

    # Top-left: data
    plot_metric(axes[0], 'data', 'Distribution of data values by issue area')

    # Top-right: performance
    plot_metric(axes[1], 'performance', 'Distribution of performance values by issue area')

    # Bottom-left: accessibility (with dotted horizontal lines marking bin edges)
    plot_metric(axes[2], 'accessibility', 'Distribution of accessibility values (bins shown)')
    # draw horizontal dotted lines for bin boundaries: acc_bins[1], acc_bins[2] (interior cuts)
    # If bins are quantiles they represent thresholds on mean_accessibility; show them on value axis
    acc_cuts = acc_bins[1:-1]  # interior edges
    y_min, y_max = axes[2].get_ylim()
    for cut in acc_cuts:
        axes[2].hlines(cut, xmin=-0.5, xmax=len(issue_order)-0.5, colors='gray',
                       linestyles='dotted', linewidth=1.2, alpha=0.8)
    # annotate bin labels on the right side of plot
    for i, lab in enumerate(acc_labels):
        # place label roughly in vertical middle of each bin
        low = acc_bins[i]
        high = acc_bins[i+1]
        ypos = (low + high) / 2.0
        axes[2].text(len(issue_order)-0.4, ypos, lab, va='center', ha='left', fontsize=9, color='gray')

    # Bottom-right: urgency (with dotted horizontal lines marking bin edges)
    plot_metric(axes[3], 'urgency', 'Distribution of urgency values (bins shown)')
    urg_cuts = urg_bins[1:-1]
    y_min, y_max = axes[3].get_ylim()
    for cut in urg_cuts:
        axes[3].hlines(cut, xmin=-0.5, xmax=len(issue_order)-0.5, colors='gray',
                       linestyles='dotted', linewidth=1.2, alpha=0.8)
    for i, lab in enumerate(urg_labels):
        low = urg_bins[i]
        high = urg_bins[i+1]
        ypos = (low + high) / 2.0
        axes[3].text(len(issue_order)-0.4, ypos, lab, va='center', ha='left', fontsize=9, color='gray')

    plt.tight_layout()
    return summary, fig, axes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Optional, Tuple, Dict

def plot_issue_distributions(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    acc_bins: Optional[List[float]] = None,
    acc_labels: List[str] = ['Low','Medium','High'],
    figsize: Tuple[int,int] = (10,18),
    uniform_color: str = '#4C72B0',     # single uniform color
    show_violin: bool = True,
    mean_line_color: str = 'black',
    mean_line_width: float = 1.6,
    mean_text_offset: float = 0.03,
    title_fontsize: int = 18,
    axis_label_fontsize: int = 16,
    tick_fontsize: int = 14,
    mean_text_fontsize: int = 12,
    xtick_rotation: int = 45,
    box_alpha: float = 0.6
) -> Tuple[pd.DataFrame, plt.Figure, np.ndarray]:
    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    label_map = {
        'global': 'Global Risks',
        'earth': 'Earth System',
        'ocean': 'Ocean Challenges',
        'water': 'Water Management',
        'biodiversity': 'Biodiversity',
        'urban': 'Urban Planet',
        'comms': 'Sustainability Comm.',   # shortened
        'policy': 'Collective Decisions'
    }

    # Gather long-form data for plotting and compute per-issue counts and means
    long_rows = []
    rows = []
    counts = {m: {} for m in ['data', 'performance', 'accessibility']}
    means = {m: {} for m in ['data', 'performance', 'accessibility']}

    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'

        if not {col_data, col_perf, col_acc}.issubset(df.columns):
            continue

        data_vals = pd.to_numeric(df[col_data], errors='coerce')
        perf_vals = pd.to_numeric(df[col_perf], errors='coerce')
        acc_vals  = pd.to_numeric(df[col_acc], errors='coerce')

        for v in data_vals:
            long_rows.append({'issue': issue, 'metric': 'data', 'value': v})
        for v in perf_vals:
            long_rows.append({'issue': issue, 'metric': 'performance', 'value': v})
        for v in acc_vals:
            long_rows.append({'issue': issue, 'metric': 'accessibility', 'value': v})

        n_data = int(data_vals.dropna().shape[0])
        n_perf = int(perf_vals.dropna().shape[0])
        n_acc  = int(acc_vals.dropna().shape[0])

        mean_data = float(data_vals.mean(skipna=True)) if n_data > 0 else np.nan
        mean_perf = float(perf_vals.mean(skipna=True)) if n_perf > 0 else np.nan
        mean_acc  = float(acc_vals.mean(skipna=True)) if n_acc > 0 else np.nan

        counts['data'][issue] = n_data
        counts['performance'][issue] = n_perf
        counts['accessibility'][issue] = n_acc

        means['data'][issue] = mean_data
        means['performance'][issue] = mean_perf
        means['accessibility'][issue] = mean_acc

        rows.append({
            'issue': issue,
            'mean_data': mean_data,
            'mean_performance': mean_perf,
            'mean_accessibility': mean_acc,
            'n_data': n_data,
            'n_performance': n_perf,
            'n_accessibility': n_acc
        })

    long_df = pd.DataFrame(long_rows)
    summary = pd.DataFrame(rows)

    if summary.empty:
        raise ValueError("No issue areas with required columns (data, performance, accessibility) found in DataFrame.")

    # Determine accessibility bin thresholds (kept for compatibility; not drawn)
    if acc_bins is None:
        try:
            q = summary['mean_accessibility'].quantile([0, 1/3, 2/3, 1.0]).values
            acc_bins_calc = np.unique(q)
            if len(acc_bins_calc) < 4:
                acc_bins = [summary['mean_accessibility'].min() - 1e-6,
                            summary['mean_accessibility'].median(),
                            summary['mean_accessibility'].mean(),
                            summary['mean_accessibility'].max() + 1e-6]
            else:
                acc_bins = q.tolist()
        except Exception:
            acc_bins = [summary['mean_accessibility'].min() - 1e-6,
                        summary['mean_accessibility'].median(),
                        summary['mean_accessibility'].mean(),
                        summary['mean_accessibility'].max() + 1e-6]

    # Set global plotting font sizes for publication
    plt.rcParams.update({
        'axes.titlesize': title_fontsize,
        'axes.labelsize': axis_label_fontsize,
        'xtick.labelsize': tick_fontsize,
        'ytick.labelsize': tick_fontsize,
        'legend.fontsize': tick_fontsize
    })

    fig, axes = plt.subplots(3, 1, figsize=figsize)
    sns.set(style='whitegrid')

    issue_order = summary['issue'].tolist()

    def display_issue_labels_for_metric(metric: str):
        labels = []
        for iss in issue_order:
            n = counts.get(metric, {}).get(iss, 0)
            disp = label_map.get(iss, iss)
            labels.append(f"{disp} (n={n})")
        return labels

    def plot_metric(ax, metric_name, title, show_mean=False):
        sub = long_df[long_df['metric'] == metric_name].copy()
        sub['issue'] = pd.Categorical(sub['issue'], categories=issue_order, ordered=True)

        if show_violin:
            # Use `color` instead of `palette` to avoid the FutureWarning
            sns.violinplot(data=sub, x='issue', y='value', ax=ax, order=issue_order,
                           inner=None, color=uniform_color, cut=0, linewidth=0.9)

        # Boxplot: use supported kwargs only
        sns.boxplot(data=sub, x='issue', y='value', ax=ax, order=issue_order,
                    whis=[5,95], width=0.12, showcaps=True, showfliers=False,
                    boxprops={'facecolor': 'white', 'alpha': box_alpha, 'edgecolor': uniform_color},
                    medianprops={'color': mean_line_color, 'linewidth':1.2},
                    whiskerprops={'color': uniform_color, 'linewidth':1.0},
                    capprops={'color': uniform_color, 'linewidth':1.0})

        ax.set_title(title, fontsize=title_fontsize)
        ax.set_xlabel('')
        ax.set_ylabel('Score', fontsize=axis_label_fontsize)

        metric_labels = display_issue_labels_for_metric(metric_name)
        ax.set_xticklabels(metric_labels, rotation=xtick_rotation, ha='right', fontsize=tick_fontsize)

        if show_mean:
            fig.canvas.draw_idle()
            y_min, y_max = ax.get_ylim()
            y_span = y_max - y_min if (y_max - y_min) != 0 else 1.0
            text_offset = mean_text_offset * y_span
            for i, iss in enumerate(issue_order):
                mean_val = means[metric_name].get(iss, np.nan)
                if np.isnan(mean_val):
                    continue
                x_center = i
                half_width = 0.25
                ax.hlines(mean_val, x_center - half_width, x_center + half_width,
                          colors=mean_line_color, linewidth=mean_line_width, zorder=4)
                txt = f"{mean_val:.1f}"
                ax.text(x_center, mean_val + text_offset, txt, ha='center', va='bottom',
                        fontsize=mean_text_fontsize, color=mean_line_color, zorder=5)

    plot_metric(axes[0], 'data', 'Data access and quality assessment by issue area', show_mean=True)
    plot_metric(axes[1], 'performance', 'Model performance assessment by issue area', show_mean=True)
    plot_metric(axes[2], 'accessibility', 'Accessibility for non-experts assessment by issue area', show_mean=True)

    plt.tight_layout()
    return summary, fig, axes
```

```python
import pandas as pd
import numpy as np

def compute_issue_mean_cuts(df, issues=None, q=[0, 1/3, 2/3, 1.0]):
    """
    Compute per-issue means and derive bin edges (quantile-based) for accessibility and urgency.
    Returns:
      - summary: DataFrame with per-issue means
      - acc_bins: list of 4 numbers [min, q1, q2, max] for accessibility means
      - urg_bins: list of 4 numbers [min, q1, q2, max] for urgency means
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    rows = []
    for issue in issues:
        col_acc = f'{issue}.accessibility'
        col_urg = f'{issue}.urgency'
        # skip if missing both
        if col_acc not in df.columns or col_urg not in df.columns:
            continue

        mean_acc = pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True)
        mean_urg = pd.to_numeric(df[col_urg], errors='coerce').mean(skipna=True)
        rows.append({'issue': issue, 'mean_accessibility': mean_acc, 'mean_urgency': mean_urg})

    summary = pd.DataFrame(rows).set_index('issue')

    if summary.empty:
        raise ValueError("No issues with accessibility/urgency columns found.")

    # Quantile-based bins on the per-issue means
    try:
        acc_q = summary['mean_accessibility'].quantile(q).values.tolist()
        # ensure unique ascending edges; if not enough unique values, fallback below
        if len(np.unique(acc_q)) < 4:
            raise ValueError("Not enough unique quantiles for accessibility; fallback.")
        acc_bins = acc_q
    except Exception:
        acc_bins = [
            summary['mean_accessibility'].min() - 1e-6,
            summary['mean_accessibility'].median(),
            summary['mean_accessibility'].mean(),
            summary['mean_accessibility'].max() + 1e-6
        ]

    try:
        urg_q = summary['mean_urgency'].quantile(q).values.tolist()
        if len(np.unique(urg_q)) < 4:
            raise ValueError("Not enough unique quantiles for urgency; fallback.")
        urg_bins = urg_q
    except Exception:
        urg_bins = [
            summary['mean_urgency'].min() - 1e-6,
            summary['mean_urgency'].median(),
            summary['mean_urgency'].mean(),
            summary['mean_urgency'].max() + 1e-6
        ]

    return summary.reset_index(), acc_bins, urg_bins

# Example usage:
# summary_means, acc_bins, urg_bins = compute_issue_mean_cuts(df_ass)
# print("Accessibility bin edges (quantiles of per-issue means):", acc_bins)
# print("Urgency bin edges (quantiles of per-issue means):", urg_bins)
# print(summary_means)  # per-issue mean values for accessibility and urgency
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from typing import List, Optional, Tuple, Dict

def plot_issue_area_means_for_ppt(
    df: pd.DataFrame,
    issues: Optional[List[str]] = None,
    annotate: bool = True,
    figsize: Tuple[int,int] = (16,8),
    title_fontsize: int = 20,
    label_fontsize: int = 16,
    tick_fontsize: int = 14,
    legend_fontsize: int = 14,
    annotation_fontsize: int = 12
) -> Tuple[pd.DataFrame, plt.Figure, Tuple[plt.Axes, plt.Axes]]:
    """
    Create two scatter subplots:
      - left: mean_data vs mean_performance
      - right: mean_accessibility vs mean_urgency

    Both subplots use simple markers (no color/size mapping).
    Axes ranges for both x and y are fixed to [0, 5].

    Returns summary DataFrame, fig, (ax_left, ax_right).
    """

    if issues is None:
        issues = ['global','earth','ocean','water','biodiversity','urban','comms','policy']

    # Build summary
    rows = []
    for issue in issues:
        col_data = f'{issue}.data'
        col_perf = f'{issue}.performance'
        col_acc  = f'{issue}.accessibility'
        col_urg  = f'{issue}.urgency'

        if not {col_data, col_perf, col_acc, col_urg}.issubset(df.columns):
            continue

        rows.append({
            'issue': issue,
            'mean_data': pd.to_numeric(df[col_data], errors='coerce').mean(skipna=True),
            'mean_performance': pd.to_numeric(df[col_perf], errors='coerce').mean(skipna=True),
            'mean_accessibility': pd.to_numeric(df[col_acc], errors='coerce').mean(skipna=True),
            'mean_urgency': pd.to_numeric(df[col_urg], errors='coerce').mean(skipna=True)
        })

    summary = pd.DataFrame(rows)
    if summary.empty:
        raise ValueError("No issue areas with all four required columns found in DataFrame.")

    # Plot setup
    sns.set(style='whitegrid')
    plt.rcParams.update({
        'font.size': tick_fontsize,
        'axes.titlesize': title_fontsize,
        'axes.labelsize': label_fontsize,
        'xtick.labelsize': tick_fontsize,
        'ytick.labelsize': tick_fontsize,
        'legend.fontsize': legend_fontsize
    })

    fig, (ax_l, ax_r) = plt.subplots(1, 2, figsize=figsize, sharey=False)

    marker_kwargs = dict(s=120, color='C0', edgecolor='k', linewidth=0.8, alpha=0.95)

    # Left plot: data vs performance
    ax_l.scatter(summary['mean_data'], summary['mean_performance'], **marker_kwargs)
    if annotate:
        for _, r in summary.iterrows():
            ax_l.text(r['mean_data'], r['mean_performance'], ' ' + r['issue'],
                      verticalalignment='center', fontsize=annotation_fontsize)

    ax_l.set_xlabel('Mean data', fontsize=label_fontsize)
    ax_l.set_ylabel('Mean performance', fontsize=label_fontsize)
    ax_l.set_title('Data vs Performance', fontsize=title_fontsize)
    ax_l.grid(alpha=0.25)
    for spine in ax_l.spines.values():
        spine.set_linewidth(1.0)
    ax_l.set_xlim(0, 5)
    ax_l.set_ylim(0, 5)

    # Right plot: accessibility vs urgency
    ax_r.scatter(summary['mean_accessibility'], summary['mean_urgency'], **marker_kwargs)
    if annotate:
        for _, r in summary.iterrows():
            ax_r.text(r['mean_accessibility'], r['mean_urgency'], ' ' + r['issue'],
                      verticalalignment='center', fontsize=annotation_fontsize)

    ax_r.set_xlabel('Mean accessibility', fontsize=label_fontsize)
    ax_r.set_ylabel('Mean urgency', fontsize=label_fontsize)
    ax_r.set_title('Accessibility vs Urgency', fontsize=title_fontsize)
    ax_r.grid(alpha=0.25)
    for spine in ax_r.spines.values():
        spine.set_linewidth(1.0)
    ax_r.set_xlim(0, 5)
    ax_r.set_ylim(0, 5)

    # Tight layout
    fig.tight_layout()

    return summary, fig, (ax_l, ax_r)
```
